### Impact of Scaling on distance Dimensionality Reduction 

Distance-based methods such as **PCA, t-SNE, and UMAP** build low-dimensional representations based on the distances between points.

- **Without scaling** → one feature with a larger numerical range can dominate all the distances, causing the dimensionality reduction to capture only that feature. This results in misleading patterns.  

- **With scaling** → all features contribute equally to the distance calculation, ensuring that the dimensionality reduction produces a more accurate and balanced representation of the data.

#### t-SNE Implementation

```python
from sklearn.manifold import TSNE
```

code for finding out best perplexity using nearest neighbour and kl divergence method

```python

from sklearn.neighbors import NearestNeighbors
def neighborhood_preservation(X_high, X_low, k=10):
    nn_high = NearestNeighbors(n_neighbors=k).fit(X_high)
    nn_low  = NearestNeighbors(n_neighbors=k).fit(X_low)
    
    high_idx = nn_high.kneighbors(return_distance=False)
    low_idx  = nn_low.kneighbors(return_distance=False)
    
    overlap = np.mean([
        len(set(high_idx[i]).intersection(set(low_idx[i]))) / k
        for i in range(len(X_high))
    ])
    return overlap

perplexities = np.arange(5, 201, 15)
kl_values = []
np_scores = []

for perp in perplexities:
    tsne = TSNE(n_components=2,perplexity=perp, random_state=42)
    X_tsne = tsne.fit_transform(x_df_scaled)
    kl = tsne.kl_divergence_
    np_score = neighborhood_preservation(x_df_scaled, X_tsne, k=10)
    kl_values.append(kl)
    np_scores.append(np_score)
    

plt.figure(figsize=(10, 5))
plt.plot(perplexities, kl_values, marker='o', color='red', linewidth=2)
plt.title('t-SNE KL Divergence vs Perplexity (5–200)', fontsize=14)
plt.xlabel('Perplexity', fontsize=12)
plt.ylabel('KL Divergence', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(perplexities, np_scores, marker='x', color='blue', linewidth=2)
plt.title('t-SNE Neighborhood Preservation vs Perplexity (5–200)', fontsize=14)
plt.xlabel('Perplexity', fontsize=12)
plt.ylabel('Neighborhood Preservation Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

### Choice of Perplexity

Based on the evaluation graphs of **KL Divergence** and **Neighborhood Preservation**:

- **KL Divergence** measures how well the **global + average similarity structure** of the data is preserved in the low-dimensional embedding.  
- **Neighborhood Preservation** measures how well the **local structure (nearest neighbors)** is preserved.  

Since a good t-SNE embedding requires a balance between **global** and **local** preservation,  
we choose **perplexity = 50**.  

At this value:
- KL divergence is reasonably low (global structure is preserved),  
- Neighborhood preservation is also high (local clusters are maintained).  

➡️ Thus, perplexity = 50 gives the best balance between **global and local structure** for our dataset.

##### Isomap Implementation

```python
from sklearn.manifold import Isomap
from sklearn.metrics import pairwise_distances
```

```python
neighbourhood=np.arange(5,200,10)
reconstruct_error=[]
for neigh in neighbourhood:
    isomap=Isomap(n_components=2,n_neighbors=neigh)
    data=isomap.fit_transform(x_df_scaled)
    G_high = isomap.dist_matrix_
    G_high = G_high[np.isfinite(G_high)] 

    G_low = pairwise_distances(data)
    G_low = G_low[np.isfinite(G_low)]

    stress = np.mean((G_high - G_low)**2)
    reconstruct_error.append(stress)

plt.figure(figsize=(8,5))
plt.plot(neighbourhood, reconstruct_error, marker='o')
plt.title("Isomap Reconstruction Error vs Number of Neighbors")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Reconstruction Error")
plt.grid(True)
plt.show()    
```

##### Fundamental Difference between Isomap and t-SNE (Global vs. Local Structure Preservation)

Isomap and t-SNE are both nonlinear dimensionality reduction techniques, but they differ fundamentally in the way they preserve the structure of high-dimensional data when mapping it to a lower-dimensional space. **Isomap** focuses on preserving the *global structure* of the data. It does this by constructing a neighborhood graph based on the shortest path (geodesic) distances between all data points, which captures the overall shape or manifold of the dataset. Once these global distances are computed, Isomap applies Multidimensional Scaling (MDS) to embed the data in a low-dimensional space while maintaining the relative distances between all points as much as possible. As a result, Isomap performs well when the data lies on a smooth and continuous manifold, such as a curved surface (e.g., a "Swiss roll").

In contrast, **t-SNE** (t-Distributed Stochastic Neighbor Embedding) prioritizes preserving the *local structure* of the data. Instead of maintaining global distances, t-SNE focuses on modeling pairwise similarities between nearby points. It converts distances in the high-dimensional space into conditional probabilities representing how likely one point is to be a neighbor of another. The same is done in the low-dimensional space, and t-SNE minimizes the divergence between the two probability distributions. This means that points that are close in high-dimensional space remain close in the low-dimensional map, effectively forming well-separated clusters. However, t-SNE does not preserve large-scale (global) relationships between clusters, which may appear arbitrarily far apart or closer than they truly are. In summary, **Isomap preserves the global geometry of the manifold**, while **t-SNE preserves the local neighborhood structure**, often at the cost of global interpretability.

##### The Concept of the Data Manifold and its Implications in Classification

The concept of a **data manifold** comes from the idea that even though gene expression profiles or other biological data may exist in a very high-dimensional space (thousands of genes as features), the data often lies on a much lower-dimensional surface or "manifold" within that space. This manifold represents the intrinsic structure of the data: continuous variations, gradients, and underlying biological processes that shape the observed patterns.

In the Isomap plot, the data appears to spread smoothly across the embedding space without forming clear, tightly separated clusters. This suggests that the underlying manifold is **highly curved or complex**, rather than being simple and linear. A simple linear manifold would have produced clearer separation of classes, while the spread-out structure indicates that the data follows a nonlinear, curved geometry in high-dimensional space.

The **complexity of the manifold** directly relates to the difficulty of classification. When the manifold is curved and the classes overlap across this curved surface, it becomes harder to draw clear linear boundaries between categories. Traditional linear classifiers may struggle in such settings, requiring more advanced methods (nonlinear models, kernel methods, or manifold learning approaches) to capture the true decision boundaries. In 
contrast, if the manifold were simple and flat, classification would be easier because classes would be more linearly separable.

**Summary:**  
- The data manifold is the lower-dimensional surface that high-dimensional data lies on.  
- The Isomap plot suggests a **complex, curved manifold** for the gene expression data.  
- Greater manifold complexity increases the difficulty of classification, as classes may 
  overlap along nonlinear trajectories.
